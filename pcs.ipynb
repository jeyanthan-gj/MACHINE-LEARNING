```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
from sklearn.datasets import load_iris


wine = load_iris()
X = wine.data
y = wine.target


wine_df = pd.DataFrame(X, columns=wine.feature_names)
wine_df['target'] = wine.target_names[y]

X = wine_df.drop('target', axis=1)
y = wine_df['target']


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Original data shape:", X.shape)
print("Standardized data shape:", X_scaled.shape)
```

```python
print("First 5 rows of the original data:\n", X.head())
```

```python
print("First 5 rows of the standardized data:\n", pd.DataFrame(X_scaled, columns=X.columns).head())
```

```python
from sklearn.decomposition import PCA
```

```python

pca=PCA(n_components=2)
```

```python
X_pca=pca.fit_transform(X_scaled)
```

```python
print("First 5 rows of the standardized data:\n", pd.DataFrame(X_scaled, columns=X.columns).head())
```

```python

print("First 5 rows of the PCA data:\n", pd.DataFrame(X_pca, columns=['PC1', 'PC2']).head())
```

```python
import matplotlib.pyplot as plt

# Scatter plot of original standardized data (first two features)
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=pd.factorize(y)[0], cmap='viridis', edgecolor='k', s=60)
plt.xlabel(X.columns[0])
plt.ylabel(X.columns[1])
plt.title('Standardized Data (First 2 Features)')

# Scatter plot of PCA data
plt.subplot(1,2,2)
plt.scatter(X_pca[:,0], X_pca[:,1], c=pd.factorize(y)[0], cmap='viridis', edgecolor='k', s=60)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Data (2 Components)')

plt.tight_layout()
```
