### **Import packages**

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## **Load dataset**

```python

breast_cancer = load_breast_cancer()
data = load_breast_cancer()


```

## **Declare dependent and independednt variables**

```python
X = data.data
y = data.target
```

### **Decalaring and fiiting data in the model**

```python
model=LinearRegression()
model.fit(X,y)
```

### Predicting Y

```python
y_pred = model.predict(X)
```

```python
y_pred
```

### **Plotting the actual and predicted values**

```python

X_feature = X[:, 0]

plt.figure(figsize=(8,6))
plt.scatter(X_feature, y, color='blue', alpha=0.5, label="Actual")
plt.scatter(X_feature, y_pred, color='red', alpha=0.5, label="Predicted")
plt.xlabel("Feature: " + data.feature_names[0])
plt.ylabel("Target Value")
plt.title("Actual vs Predicted values for " + data.feature_names[0])
plt.legend()
plt.show()
```

### Score of the model

```python
model.score(X,y)
```

### **Importing Ridge**

```python
from sklearn.linear_model import Ridge
```

### **Declaring model and fitting data**

```python
Ridge=Ridge(alpha=1.0)
Ridge.fit(X,y)
```

## **Predicting Y**

```python
yr_pred=Ridge.predict(X)
yr_pred
```

### **Plotting the actual and predicted values**

```python
X_feature = X[:, 0]
feature_name = data.feature_names[0]

plt.figure(figsize=(8,6))
plt.scatter(X_feature, y, color='blue', alpha=0.6, label='Actual (y)')
plt.scatter(X_feature, yr_pred, color='red', alpha=0.6, label='Predicted (yr_pred)')
plt.xlabel(feature_name)
plt.ylabel("Target / Prediction")
plt.title("Ridge Regression: Actual vs Predicted")
plt.legend()
plt.show()
```

### **Score**

```python
Ridge.score(X,y)
```
